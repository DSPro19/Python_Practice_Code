```python
#Importing the modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('1.04.+Real-life+example.csv')
```

```python
raw_data.head()
```

```python
raw_data.describe(include = 'all')
```

```python
data = raw_data.drop(['Registration'], axis = 1)
data.isnull().sum()
```

```python
data_no_mv = data.dropna(axis=0)
```

```python
data_no_mv.describe(include='all')
```

```python
sns.distplot(data_no_mv['Price'])
```

```python
q = data_no_mv['Price'].quantile(0.99)
w = data_no_mv['Price'].quantile(0.001)
data_1 = data_no_mv[data_no_mv['Price']<q]
data_2 = data_1[data_1['Price']>w]
sns.distplot(data_2['Price'])
```

```python
sns.distplot(data_2['Mileage'])
```

```python
q1 = data_2['Mileage'].quantile(0.99)
w1 = data_2['Mileage'].quantile(0.001)
data_3 = data_2[data_2['Mileage']<q1]
data_4 = data_3[data_3['Mileage']>w1]
sns.distplot(data_4['Mileage'])
```

```python
data_4.describe(include='all')
```

```python
data_5 = data_4[data_4['EngineV']<6.5]
data_cleaned = data_5.reset_index(drop=True)
data_cleaned.describe(include='all')
```

```python
log_price = np.log(data_cleaned['Price'])
data_cleaned['Log_price'] = log_price
```

```python
f, ( ax1, ax2, ax3) = plt.subplots(1, 3,sharey=True, figsize=(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['Price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['Price'])
ax3.set_title('Price and Mileage')

plt.show()
```

```python
f, ( ax1, ax2, ax3) = plt.subplots(1, 3,sharey=True, figsize=(21,9))
ax1.scatter(data_cleaned['Year'],data_cleaned['Log_price'])
ax1.set_title('Log Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['Log_price'])
ax2.set_title('Log Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['Log_price'])
ax3.set_title('Log Price and Mileage')

plt.show()
```

```python
data_clean = data_cleaned.drop(['Model','EngineV','Price'], axis=1)
data_with_dummies = pd.get_dummies(data_clean, drop_first=True, dtype=int)
data_with_dummies.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variable = data_with_dummies[[ 'Year', 'Mileage', 'Log_price', 'Brand_BMW', 'Brand_Mercedes-Benz', 'Brand_Mitsubishi','Brand_Renault','Brand_Toyota','Brand_Volkswagen','Body_hatch', 'Body_other', 'Body_sedan', 'Body_vagon', 'Body_van','Engine Type_Gas','Engine Type_Other','Engine Type_Petrol']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variable.values,i) for i in range (variable.shape[1])]
vif['Features'] = variable.columns
vif
```

```python
data_with_dummies.columns.values
```

```python
cols = ['Mileage', 'Year', 'Log_price', 'Brand_BMW', 'Brand_Mercedes-Benz',
       'Brand_Mitsubishi', 'Brand_Renault', 'Brand_Toyota',
       'Brand_Volkswagen', 'Body_hatch', 'Body_other', 'Body_sedan',
       'Body_vagon', 'Body_van', 'Engine Type_Gas', 'Engine Type_Other',
       'Engine Type_Petrol']
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

```python
targets = data_preprocessed['Log_price']
inputs = data_preprocessed.drop(['Log_price'], axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
inputs_scaled = scaler.transform(inputs)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size = 0.1, random_state = 365)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets(y_train)', size = 19)
plt.ylabel('Predictions(y_hat)', size = 19)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
reg.score(x_train, y_train)
```

```python
sns.distplot(y_train - y_hat)
plt.title("PDF", size = 19)
```

```python
##Weights And Bias
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns = ['Features'])
reg_summary['Weights']  = reg.coef_
reg_summary
```

```python
y_hat_test = reg.predict(x_test)
plt.scatter(y_test, y_hat_test, alpha = 0.2)
plt.xlabel('Targets(y_test)', size = 19)
plt.ylabel('Predictions(y_hat_test)', size = 19)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns = ['Prediction'])
df_pf.head()
```

```python
y_test = y_test.reset_index(drop = True)
df_pf['Target'] = np.exp(y_test)
y_test.head()
df_pf
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
df_pf['Difference%'] = np.absolute(df_pf['Residual']/ df_pf['Target']*100)
df_pf.describe()
```
